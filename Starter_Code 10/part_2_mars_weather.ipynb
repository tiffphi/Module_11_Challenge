### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
mars_month_counts = mars_temp_df["month"].value_counts().sort_index()
mars_month_counts
```

```python
# 2. How many Martian days' worth of data are there?
mars_temp_df['sol'].count()
```

```python
# 3. What is the average low temperature by month?
month = mars_temp_df.groupby(['month'])
avg_low_temp = month["min_temp"].mean()
avg_low_temp
```

```python
# Plot the average temperature by month
import numpy as np

x_axis = np.arange(1, len(month)+1)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis)
plt.bar(x_axis, avg_low_temp, color="b", align="center")
plt.title("Average Temperature by Month")
plt.xlabel("month")
plt.ylabel("Temperature in Celsius")
plt.show()
```

```python
month = mars_temp_df.groupby(['month'])
asc_avg_low_temp = month["min_temp"].mean().sort_values()
asc_avg_low_temp
```

```python
# Identify the coldest and hottest months in Curiosity's location

x_axis = asc_avg_low_temp.index
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis)
plt.bar(x_axis, asc_avg_low_temp, color="b", align="center")
plt.title("Average Temperature by Month")
plt.xlabel("month")
plt.ylabel("Temperature in Celsius")

plt.show()

#The image has the correct information however, it is not visualized in ascending
#ordrer for temp, it's still in ascending order by month(x-axis)
```

```python
coolest_month = asc_avg_low_temp.idxmin()
print(f'The Coolest Month on Mars is month number {coolest_month}')
```

```python
hottest_month = asc_avg_low_temp.idxmax()
print(f'The Hottest Month on Mars is month number {hottest_month}')
```

```python
# 4. Average pressure by Martian month
month = mars_temp_df.groupby(['month'])
avg_pressure = month["pressure"].mean()
avg_pressure
```

```python
#The image displays the same data that's being requested, it just is in ascending 
#order for the months, instead of ascending order for the atmospheric pressure.
x_axis = avg_pressure.index
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis)
plt.bar(x_axis, avg_pressure, color="b", align="center")
plt.title("Average Pressure by Month")
plt.xlabel("month")
plt.ylabel("Atmospheric Pressure")
plt.show()
```

```python
#Which months have the lowest and the highest atmospheric pressure on Mars? 
lowest_pressure = avg_pressure.idxmin()
print(f'The Month with the lowest atmospheric pressure on Mars is month number {lowest_pressure}')
highest_pressure = avg_pressure.idxmax()
print(f'The Month with the highest atmospheric pressure on Mars is month number  {highest_pressure}')
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

earth_days = mars_temp_df['terrestrial_date'].value_counts()
x_axis = np.arange(len(earth_days)) 
mars_low_temp = mars_temp_df['min_temp']
mars_low_temp
plt.plot(x_axis, mars_low_temp, color="blue")

plt.xlabel("Number of Terrestrial Days")
plt.ylabel("Minimum Temperature")
plt.show()
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
